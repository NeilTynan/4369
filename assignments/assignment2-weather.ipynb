```python
# Importing Libraries
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Loading in the CSV File
df = pd.read_csv('weatherreadings1.csv')
```

```python
# Define Data
dry = df['dryBulbTemperature_Celsius']
wet = df['wetBulbTemperature_Celsius']
date = df['reportStartDateTime']
date2 = np.asarray(date, dtype='datetime64[s]')

plt.plot(date2, dry, 'r', label = 'Dry Bulb')
plt.plot(date2, wet, 'b', label = 'Wet Bulb')

plt.tick_params(axis='x', labelrotation=90)
plt.xlabel ('Time (Month-Day-Hour)', style='italic')
plt.ylabel ('Temperature (*C)', style='italic')
plt.title('Dry Bulb vs Wet Bulb Temperature Over Time', fontweight='bold', style='italic', family='monospace') # Add Title


plt.legend() # Add Legend
plt.show()  # display
```
